```python
import glob
import matplotlib.pyplot as plt
from ultralytics import YOLO

model = YOLO("/kaggle/input/yolo-clothes-seg/YOLO Clothes.pt")

results = model.predict(
    source="/kaggle/input/clothing-seg/test/images",  
    conf=0.5
)

num_to_show = 5  

for r in results[:num_to_show]:
    plt.figure(figsize=(6,6))
    plt.imshow(r.plot()) 
    plt.axis("off")
    plt.show()
```
